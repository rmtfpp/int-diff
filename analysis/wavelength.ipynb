DIFFRACTION FUNCTION WITH DATA MANUALLY TAKEN FROM CSV

```python
dmaxd = 75.29e-3
a = 0.16e-3
L = 0.9

def wavelength_diffraction(dataset):
  wls = []
  for i in dataset:
    wls.append((a*(i[1] - dmaxd))/(L*i[0]))
  wl = sum(wls) / len(wls)
  return wl
```

```python
# Ks and respective Ys. Data of local minima manually extracted from csv

ddata = [
    [1, 78.96e-3],
    [2, 82.58e-3],
    [3, 86.23e-3],
    [4, 89.82e-3],
    [-1, 71.66e-3],
    [-2, 67.92e-3],
    [-3, 64.23e-3],
    [-4, 60.71e-3]
]

print(wavelength_diffraction(ddata))
```

```python
# error percentage on declared wavelength value

print(((650e-9 - wavelength_diffraction(ddata))/(650e-9)) * 100)
```

INTERFERENCE


```python
imaxd = 75.91e-3
d = 0.5e-3
L = 0.9

def wavelength_interference(dataset):
  wls = []
  for i in dataset:
    wls.append((d*(i[1] - imaxd))/(L*i[0]))
  wl = sum(wls) / len(wls)
  return wl
```

```python
# Ks and respective Ys. Data of local maxima manually extracted from csv

idata = [
    [1, 77.08e-3],
    [2, 78.27e-3],
    [3, 79.45e-3],
    [-1, 74.76e-3],
    [-2, 73.58e-3],
    [-3, 72.48e-3]
]

wavelength_interference(idata)
```

```python
# error percentage on declared wavelength value

((650e-9 - wavelength_interference(idata))/(650e-9)) * 100
```
